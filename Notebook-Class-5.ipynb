```python
word = "a1b2c3"
letters = list(filter(str.isalpha, word))
numbers = list(filter(str.isdigit, word))

print(letters)
print(numbers)

for letter, number in zip(letters, numbers):
  print(letter, number)
```

map

```python
def make_square(x):
  return x*x
```

```python
num_list = [2,4,1,3]
square_list = list(map(make_square, num_list))
print(square_list)
```

```python
words = ['python', 'java', 'c++']
upper_words = list(map(str.upper, words))
print(upper_words)
```

lambda function

```python
num_list = [1, 6, 7, 9, 0, 8, 4]

even_list = list(filter(lambda x: x%2==0, num_list))
print(even_list)
```

```python
square_list = list(map(lambda x: x*x, num_list))
print(square_list)
```

```python
a = [1, 2, 3]
b = [10, 20, 30]
c = list(map(lambda x, y: x+y, a, b))
print(c)
```

Introduction to OOP

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def describe(self):
    print(f"My name is {self.name}, my age is {self.age}, my phone number is {self.phone_number}")

  def set_phone_number(self, phone_number):
    self.phone_number = phone_number
```

```python
p1 = Person("Rahim", 25)
p1.set_phone_number("0123456789")
p1.describe()
# p2 = Person("Karim", 35)
# p2.describe()
```
